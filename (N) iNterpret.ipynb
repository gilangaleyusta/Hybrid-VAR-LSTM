```python
import pandas as pd
import matplotlib.pyplot as plt

# Data Historis
data_historis = {
    'Date': [
        '2023-03-15', '2023-04-15', '2023-05-15', '2023-06-15',
        '2023-07-15', '2023-08-15', '2023-09-15', '2023-10-15',
        '2023-11-15', '2023-12-15'
    ],
    'Close': [
        6748.64, 6831.37, 6735.19, 6673.74,
        6826.16, 6899.06, 6967.65, 6861.98,
        6909.09, 7162.59
    ],
    'Kurs Dollar': [
        15300.71, 14866.71, 14810.86, 14932.00,
        15039.75, 15245.14, 15354.20, 15741.23,
        15617.41, 15513.37
    ]
}

df_historis = pd.DataFrame(data_historis)
df_historis['Date'] = pd.to_datetime(df_historis['Date'])
df_historis.set_index('Date', inplace=True)

# Data Forecasting
data_forecasting = {
    'Date': [
        '2024-01-10', '2024-02-10', '2024-03-10', '2024-04-10',
        '2024-05-10', '2024-06-10', '2024-07-10'
    ],
    'Actual Close': [
        7238.07, 7284.16, 7339.20, 7165.03,
        7154.31, 6904.72, 7257.83
    ],
    'Actual Kurs Dollar': [
        15610.82, 15665.33, 15702.61, 16100.00,
        16083.94, 16329.39, 16261.65
    ],
    'Forecast Close (VAR)': [
        7269.05, 7318.74, 7329.97, 7344.44,
        7365.40, 7392.62, 7421.29
    ],
    'Forecast Kurs Dollar (VAR)': [
        15535.73, 15561.97, 15599.68, 15647.70,
        15684.61, 15710.49, 15738.08
    ],
    'Forecast Close (LSTM Multivariate)': [
        6929.18, 6965.63, 6977.62, 6965.84, 6950.49, 6934.32, 6918.55
    ],
    'Forecast Kurs Dollar (LSTM Multivariate)': [
        15055.16, 14990.66, 14952.70, 14933.47, 14919.01, 14906.94, 14895.76
    ],
    'Forecast Close (Hybrid)': [
        7179.22, 7257.41, 7271.04, 7275.81,
        7281.15, 7289.94, 7298.88
    ],
    'Forecast Kurs Dollar (Hybrid)': [
        15742.16, 15751.23, 15755.43, 15774.28,
        15789.50, 15794.96, 15801.59
    ]
}

df_forecasting = pd.DataFrame(data_forecasting)
df_forecasting['Date'] = pd.to_datetime(df_forecasting['Date'])
df_forecasting.set_index('Date', inplace=True)

# Adjusting the boundary line to be placed between the historical data and forecasting data
boundary_date_adjusted = df_forecasting.index.min() - pd.Timedelta(days=1)
```

```python
# Re-plotting the Close data with adjusted boundary line
plt.figure(figsize=(12, 6))
plt.plot(df_historis.index, df_historis['Close'], label='Historical Close', color='blue', linestyle='-', marker='o')
plt.plot(df_forecasting.index, df_forecasting['Actual Close'], label='Actual Close', color='green', linestyle='-', marker='o')
plt.plot(df_forecasting.index, df_forecasting['Forecast Close (VAR)'], label='Forecast Close (VAR)', color='red', linestyle='-', marker='o')
plt.plot(df_forecasting.index, df_forecasting['Forecast Close (LSTM Multivariate)'], label='Forecast Close (LSTM)', color='orange', linestyle='-', marker='o')
plt.plot(df_forecasting.index, df_forecasting['Forecast Close (Hybrid)'], label='Forecast Close (Hybrid)', color='purple', linestyle='-', marker='o')
plt.axvline(boundary_date_adjusted, color='black', linestyle='--', linewidth=1, label='Forecasting Start')
plt.xlabel('Date')
plt.ylabel('Close (Rp)')
plt.title('Comparison of Different Methods - Close IHSG')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Re-plotting the Kurs Dollar data with adjusted boundary line
plt.figure(figsize=(12, 6))
plt.plot(df_historis.index, df_historis['Kurs Dollar'], label='Historical Kurs Dollar', color='blue', linestyle='-', marker='o')
plt.plot(df_forecasting.index, df_forecasting['Actual Kurs Dollar'], label='Actual Kurs Dollar', color='green', linestyle='-', marker='o')
plt.plot(df_forecasting.index, df_forecasting['Forecast Kurs Dollar (VAR)'], label='Forecast Kurs Dollar (VAR)', color='red', linestyle='-', marker='o')
plt.plot(df_forecasting.index, df_forecasting['Forecast Kurs Dollar (LSTM Multivariate)'], label='Forecast Kurs Dollar (LSTM)', color='orange', linestyle='-', marker='o')
plt.plot(df_forecasting.index, df_forecasting['Forecast Kurs Dollar (Hybrid)'], label='Forecast Kurs Dollar (Hybrid)', color='purple', linestyle='-', marker='o')
plt.axvline(boundary_date_adjusted, color='black', linestyle='--', linewidth=1, label='Forecasting Start')
plt.xlabel('Date')
plt.ylabel('Kurs Dollar (Rp)')
plt.title('Comparison of Different Methods - Kurs Dollar (USD/ IDR)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
